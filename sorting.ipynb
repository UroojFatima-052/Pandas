## Sorting
Sorting means arranging data in a specific order
sort_values() and sort_index()

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'Name' : ['Zaryab', 'Laiba', 'Sharukh', 'Bisma', 'Muskan'],
    'Age' : [23, 45, 12, 56, 78],
    'Salary' : [20000, 30000, 300, 40000, 25000],
    'Gender' : ['M', 'F', 'M', 'F', 'F']

}, index=[2, 4, 1, 5, 3])

df
```

```python
df.sort_values(by='Age', ascending=False)
```

```python
df.sort_index()
```
